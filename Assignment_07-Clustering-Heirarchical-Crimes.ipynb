## Assignment-07-Clustering-Hierarchical(Crimes)

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# import dataset
crime=pd.read_csv(R"\Users\ammaurya\Desktop\Data Science\Assignments\7-Clustering\crime_data.csv")
crime
```

```python
crime.info()
```

```python
# Normalize heterogenous numerical data using z-score (x-mean/std) or custom defined function 
# Normalization function - here custom defined
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
crime_norm=norm_func(crime.iloc[:,1:])
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(crime_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(crime_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
crime['clustersid']=hclusters.labels_
crime
```

```python
crime.groupby('clustersid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clustersid'],crime['Rape'], c=hclusters.labels_) 
```
